```python
from supabase import create_client, Client

# Initialize Supabase client
supabase_url = 'https://hpogngdwousevnyrnmew.supabase.co'
supabase_key = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJpc3MiOiJzdXBhYmFzZSIsInJlZiI6Imhwb2duZ2R3b3VzZXZueXJubWV3Iiwicm9sZSI6InNlcnZpY2Vfcm9sZSIsImlhdCI6MTcxNDg5NDE1MiwiZXhwIjoyMDMwNDcwMTUyfQ.ZzvQboRczmpcu2AgCx8YdTXm82AsFerZYe8Nzkx7JbI'
supabase: Client = create_client(supabase_url, supabase_key)

# Duplicate products with live IDs and update prices
def duplicate_products_and_update_prices(test_products, live_products):
    # Step 1: Update prices with live product IDs
    for test_product in test_products:
        matching_live_product = next((p for p in live_products if p['name'] == test_product['name']), None)
        if matching_live_product:
            # Sort prices by unit amount to ensure higher price is for the year
            test_product['prices'].sort(key=lambda x: x['unit_amount'])
            matching_live_product['prices'].sort(key=lambda x: x['unit_amount'])
            
            for test_price, live_price in zip(test_product['prices'], matching_live_product['prices']):
                print(f"\nUpdating price for product name: {test_product['name']}")
                print(f"Test mode price ID: {test_price['id']} with interval {test_price['interval']}")
                print(f"Live mode price ID: {live_price['id']} with interval {live_price['interval']}")

                # Ensure that intervals match before updating
                if test_price['interval'] == live_price['interval']:
                    # Check if price with live ID already exists in Supabase
                    response = supabase.table('prices').select('id').eq('id', live_price['id']).execute()
                    if response.data:
                        print(f"Price with ID {live_price['id']} already exists. Skipping update.")
                        continue

                    price_update = {
                        'id': live_price['id'],
                        'product_id': matching_live_product['id']
                    }
                    # Update price in Supabase
                    response = supabase.table('prices').update(price_update).eq('id', test_price['id']).execute()
                    if not response.data:
                        print(f"Error updating price {test_price['id']} to {live_price['id']}: {response}")
                    else:
                        print(f"Successfully updated price {test_price['id']} to {live_price['id']}")
                else:
                    print(f"Intervals do not match for prices: {test_price['interval']} and {live_price['interval']}")

    # Step 2: Duplicate test products with live product IDs
    for test_product in test_products:
        matching_live_product = next((p for p in live_products if p['name'] == test_product['name']), None)
        if matching_live_product:
            print(f"Processing product: {test_product['name']}")
            print(f"Test mode product ID: {test_product['id']}")
            print(f"Live mode product ID: {matching_live_product['id']}")

            # Check if product with live ID already exists in Supabase
            response = supabase.table('products').select('id').eq('id', matching_live_product['id']).execute()
            if response.data:
                print(f"Product with ID {matching_live_product['id']} already exists. Skipping duplication.")
                continue

            # Fetch test product details from Supabase
            response = supabase.table('products').select('*').eq('id', test_product['id']).execute()
            if not response.data:
                print(f"Error fetching product details for {test_product['id']}: {response}")
                continue
            test_product_details = response.data[0]

            # Insert duplicate product with live product ID
            product_data = {
                'id': matching_live_product['id'],
                'active': test_product_details['active'],
                'name': test_product_details['name'],
                'description': test_product_details['description'],
                'image': test_product_details['image'],
                'metadata': test_product_details['metadata'],
                'features': test_product_details['features'],
                'data_sources_limit': test_product_details['data_sources_limit'],
                'questions_limit': test_product_details['questions_limit'],
            }
            response = supabase.table('products').insert(product_data).execute()
            if not response.data:
                print(f"Error duplicating product {test_product['id']} with {matching_live_product['id']}: {response}")
            else:
                print(f"Successfully duplicated product {test_product['id']} with {matching_live_product['id']}")

    # Step 3: Delete test mode products
    for test_product in test_products:
        print(f"Deleting test mode product: {test_product['name']}")
        response = supabase.table('products').delete().eq('id', test_product['id']).execute()
        if not response.data:
            print(f"Error deleting test mode product {test_product['id']}: {response}")
        else:
            print(f"Successfully deleted test mode product {test_product['id']}")
```
